# K Means Clustering

K Means Clustering is an unsupervised learning algorithm i.e, it works on unlabelled data used to group similar data into K clusters

K-means does this in 3 main steps:
1. Choose K (number of clusters)
2. Pick K points as centres or (centroids) randomly
3. Repeat until stable:
* Assign each point to the nearest center (formaing clusters)
* Move the centre to the average position of the points in that cluster

* There is no easy way of choosing the best "K" value.. no of clusters
* One way is the lebow method
* First of all, compute the sum of squared error(SSE) for some values of K (eg 2,4,6,8 etc)
* The SSE is defined as the sum of squared distance between each member of the cluster and its centroid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#K means clustering is an unsupervised learning algorithm which means we are not going to predict anything
#but instead we are going to just arrange the data into groups
```

```python
from sklearn.datasets import make_blobs
```

```python
data= make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
#make_blobs creates a dataset of points grouped into clusters
#200 data points and each point has 2 features
#centers=4.. creates 4 clusters
#cluster_std=1.8.. Std deviation of clusters/ How sprad out the points are
```

```python
data[0].shape  #200 rows and 2 columns
```

```python
#create a scatter plot
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
#data[:,0].. all x coordinates
#data[:,1].. all y coordinates
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans= KMeans(n_clusters=8)   #we are going to organize the data into 8 clusters only
```

```python
kmeans.fit(data[0])
#here we are not trying to predict anything... but rather, fit the data in groups
```

```python
kmeans.cluster_centers_
#after fitting the data into groups.. calculating the centroid of each cluster
```

```python
kmeans.labels_
#lables_ store the cluster label assigned to each data point
```

```python
fig, (ax1,ax2)= plt.subplots(1,2, sharey=True, figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='Set1')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')

#creating a subplot of 1 row and 2 columns in which they share y
#1st plot colors the data based on the training of the model
#2nd plot colors the data based on the actual values assigned to them
```
