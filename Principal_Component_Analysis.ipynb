# Principal Component Analysis

Principal Component Analysis is a technique used to reduce the number of features in your data while keeping as much as information as possible

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
#we are gonna find out which components are most important that explains the variance of the dataset
```

```python
df= pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target']
```

```python
cancer['target_names']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data= scaler.transform(df)
```

## Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)   #we are gonna find the 2 important components that are very much required for visualising the data
```

```python
pca.fit(scaled_data)
```

```python
x_pca= pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])    #c determines, based on which value they color
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend()
```

```python
pca.components_
```

```python
df_comp= pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
